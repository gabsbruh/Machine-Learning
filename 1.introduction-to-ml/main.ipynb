```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

iris dataset


```python
from sklearn.datasets import load_iris
# load the famous iris data
irisRaw = load_iris()
```

```python
# read iris.data into a pandas DataFrame (df), including column names
iris = pd.DataFrame(data= np.c_[irisRaw['data'], irisRaw['target']],
                     columns= irisRaw['feature_names'] + ['target'])
```

8. Find the minimum and maximum petal length (column 3 - petal length). Give the index of each of these values.

```python
min_petal_length = iris[iris["petal length (cm)"] == iris["petal length (cm)"].min()]
print("min_petal_length:")
print(min_petal_length)

max_petal_length = iris[iris["petal length (cm)"] == iris["petal length (cm)"].max()]
print("max_petal_length:")
print(max_petal_length)
```

9. Calculate the standard deviation for each column

```python
col_list = iris.values
x = 0
for col in iris:
  print(f"standrad deviatoin of {col}: {iris[col].std()}")
  x += 1
```

10. Identify rows for which the length of the flower cup (sepal length) is greater than the average length (the whole set)


```python
sepal_len_average = iris["sepal length (cm)"].mean()
print("average: sepal_len_average")
above_average = iris[(iris["sepal length (cm)"] > sepal_len_average)]
above_average
```

11. Display histograms for individual parameters

```python
import matplotlib.pyplot as plt

for col in iris:
    plt.hist(iris[col], 50, density=True, facecolor='g', alpha=0.75)
    plt.xlabel(col)
    plt.ylabel('Density')
    plt.title(f'Histogram of {col}')
    plt.grid(True)
    plt.show()
```
